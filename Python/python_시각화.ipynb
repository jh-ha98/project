```python
import csv
import numpy
import pandas as pd
import matplotlib.pyplot as plt

csv_data=pd.read_csv('c:/my_data/해외취업.csv', encoding='euc-kr')
csv_data
```

```python
나라=csv_data.filter(['일본','아랍에미리트','연도'])
나라df=나라.groupby('연도').sum()
나라df
```

```python
a_txt=list(나라df.index)

for i, v in enumerate(a_txt):
    a_txt[i] = "%d년" % v
    
a_txt

나라copy=나라df.copy();

idx=list(나라copy.index)

indexDict = {}

for i, v in enumerate(a_txt):
    bf = idx[i];
    indexDict[bf] = v
    
print(indexDict)

xpos = numpy.arange(len(나라copy.index))
xpos
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 16
plt.rcParams["figure.figsize"] = (20, 10)

나라df.plot()
plt.grid()
plt.legend()
plt.title("나라별 취업증가세 비교") 
plt.xlabel("나라")
plt.ylabel("계")
plt.xticks(나라df.index, a_txt)
plt.show()
```

```python
해외취업=csv_data.filter(['연도','계'])
해외취업df=해외취업.groupby('연도').sum()
해외취업df
```

```python
해외취업df['계'].plot.bar(color='#87ceeb', rot=0, width=0.7)
plt.rc('font',family='Malgun Gothic')
plt.grid()
plt.title('연도별 해외취업 분포')
plt.xlabel('연도')
plt.ylabel('계')

for x,y in enumerate(list(해외취업df['계'])):
    txt= "%d건" % y
    plt.text(x, y, txt, fontsize=14, color='#ff0000', 
                    horizontalalignment='center', verticalalignment='bottom')

plt.show()
```

```python
import csv
import numpy
from pandas import DataFrame
import matplotlib.pyplot as plt

교통사고=pd.read_csv('c:/my_data/졸음운전.csv', encoding='euc-kr')
교통사고
```

```python
교통사고tmp=교통사고.filter(['년도','사고(건)','사망(명)','부상(명)'])
교통사고df=교통사고tmp.groupby(['년도']).sum()
교통사고df
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 16
plt.rcParams["figure.figsize"] = (20, 10)

교통사고df.plot()
plt.grid()
plt.legend()
plt.title("년도별 교통사고 발생현황")
plt.xlabel("년도")
plt.ylabel("교통사고 수")
plt.show()
```

```python
교통사고평균=교통사고df.mean()
교통사고평균
```

```python
교통사고평균df = DataFrame(교통사고평균, columns=['교통사고'])
교통사고평균df
```

```python
def make_autopct(비율):
    global 교통사고평균df    
    print(비율)
    
    합계 = 교통사고평균df['교통사고'].sum()
    수치 = int(round(비율*합계/100.0))      
    str = "%0.1f%%\n(%d건)" % (비율, 수치)  
    return str

교통사고평균df['교통사고'].plot.pie(labels=None,
                                 textprops={'color':"#ffffff", 'fontsize': 18},
                                 colors=['#ffa500', '#000080', '#2e8b57'], 
                                 explode=[0.0, 0.1, 0.0], 
                                 autopct=make_autopct, 
                                 shadow=True, 
                                 startangle=90)
plt.ylabel(None)

plt.legend(labels=list(교통사고평균df.index), bbox_to_anchor=(0.95, 0.625))

plt.show()
```
